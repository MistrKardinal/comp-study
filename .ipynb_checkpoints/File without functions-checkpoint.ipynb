```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def diff_eq(z, y, g2, g3):
    """
     : y' = sqrt(4*y - g2*y - g3)
        
    """
    expr = 4*y[0]**3 - g2*y[0] - g3
    # ,     
    if expr < 0:
        return [0]  #     
    return [np.sqrt(expr)]

# 
g2 = (4*(N+1)**2)/3
g3 = -(8*(N+1)**3)/27 + N
y0 = [1]  #  
z_span = [0.0001, 0.1]  #  

# 
sol = solve_ivp(diff_eq, z_span, y0, args=(g2, g3), 
                dense_output=True, rtol=1e-8, atol=1e-8)

# 
z_plot = np.linspace(z_span[0], z_span[1], 1000)
y_plot = sol.sol(z_plot)[0]

plt.figure(figsize=(10, 6))
plt.plot(z_plot, N/(y_plot - (N+1)/3 + 1), 'b-', label='y(z)')
plt.xlabel('z')
plt.ylabel('y(z)')
plt.title(f': (y\') = 4y - {g2}y - {g3}')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.special import ellipj
from scipy.optimize import root
import matplotlib.pyplot as plt

def weierstrass_invariants_to_parameters(g2, g3):
    """
          g2, g3
      e1, e2, e3   m
    """
    #   : 4t^3 - g2*t - g3 = 0
    coefficients = [4, 0, -g2, -g3]
    roots = np.roots(coefficients)
    roots = np.sort(roots[np.isreal(roots)].real)
    
    if len(roots) < 3:
        raise ValueError("    ")
    
    e1, e2, e3 = roots[2], roots[1], roots[0]  # e1 > e2 > e3
    
    #   m
    m = (e2 - e3) / (e1 - e3)
    
    return e1, e2, e3, m

def weierstrass_p(z, g2, g3):
    """
       (z; g2, g3)
    """
    try:
        e1, e2, e3, m = weierstrass_invariants_to_parameters(g2, g3)
        
        #   
        omega = 1.0 / np.sqrt(e1 - e3)
        
        #  
        u = z * np.sqrt(e1 - e3)
        
        #    
        sn, cn, dn, phi = ellipj(u, m)
        
        return e3 + (e1 - e3) / sn**2
        
    except Exception as e:
        print(f": {e}")
        return np.nan * np.ones_like(z)

#  
g2 = (4*(N+1)**2)/3
g3 = -(8*(N+1)**3)/27 + N
z_values = np.linspace(0.0001, 0.1, 1000)  #    0

#  
e1, e2, e3, m = weierstrass_invariants_to_parameters(g2, g3)
print(f": e1 = {e1:.3f}, e2 = {e2:.3f}, e3 = {e3:.3f}")
print(f" m = {m:.3f}")

#  
y_analytic = weierstrass_p(z_values, g2, g3)

# 
plt.figure(figsize=(12, 8))
plt.plot(z_values, y_analytic, 'r-', linewidth=2, label=f'(z; {g2}, {g3})')
plt.axhline(y=e1, color='g', linestyle='--', label=f'e1 = {e1:.2f}')
plt.axhline(y=e2, color='b', linestyle='--', label=f'e2 = {e2:.2f}') 
plt.axhline(y=e3, color='m', linestyle='--', label=f'e3 = {e3:.2f}')
plt.xlabel('z')
plt.ylabel('(z)')
plt.title('  ')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

```python
def W_function_global(t, g2, g3):
    rho = np.sqrt(g2/3)
    theta = math.acos(3*np.sqrt(3)*g3/(g2**(1.5)))/3
    e1 = rho*np.cos(theta)
    e2 = rho*np.cos(theta - 2*np.pi/3)
    e3 = rho*np.cos(theta - 4*np.pi/3)
    sorted_list = [e1, e2, e3]
    sorted_list.sort()
    return W_function(t, sorted_list[2], sorted_list[1], sorted_list[0])
```

```python
def W_function(t, e1, e2, e3):
    m = (e2-e3)/(e1-e3)
    alpha = np.sqrt(e1-e3)
    sn, cn, dn, ph = scipy.special.ellipj(alpha*t, m)  
    return e3 + (e1 - e3)/sn**2
```

```python
def help_function(alpha, N):
    count = 2*N*np.log(alpha) - alpha**2
    for i in range(1,N+1):
        count -= np.log(i)
    return np.exp(count)
```

```python
def n_alpha(alpha, t):
    count = 0
    for i in range(1,300):
        g2 = (4*(i+1)**2)/3
        g3 = -(8*(i+1)**3)/27 + i
        count += help_function(alpha, i)*(i/(W_function_global(t, g2, g3) - (i+1)/3 + 1))
    return count
```
