#### Наша задача промоделировать процесс вырожденного спонтанного параметрического рассеяния (DSPDC), гамильтониан этого процесса имеет следующий вид:
$\hat{H} = H_o + H_{spdc} = \hbar w \hat{a}^{\dagger}\hat{a} + \dfrac{\hbar w}{2}\hat{b}^{\dagger}\hat{b} + i\hbar \chi^{(2)}(\hat{a}^{\dagger}\hat{b}\hat{b} - \hat{a}\hat{b}^{\dagger}\hat{b}^{\dagger})$


$H_{spdc} = i\hbar \chi^{(2)}(\hat{a}^{\dagger}\hat{b}\hat{b} - \hat{a}\hat{b}^{\dagger}\hat{b}^{\dagger}) = -i\dfrac{\hbar \chi^{(2)}}{2\sqrt{2}}((\hat{x_a} + \dfrac{d}{dx_a})(\hat{x_b^2} + \dfrac{d^2}{dx_b^2} - \hat{1_b} - 2x_b\dfrac{d}{dx_b}) - (\hat{x_a} - \dfrac{d}{dx_a})(\hat{x_b^2} + \dfrac{d^2}{dx_b^2} + \hat{1_b} + 2x_b\dfrac{d}{dx_b})) = -i\dfrac{\hbar \chi^{(2)}}{2\sqrt{2}}(2\dfrac{d}{dx_a}(\hat{x_b^2} + \dfrac{d^2}{dx_b^2}) - 2\hat{x_a}(\hat{1_b} + 2x_b\dfrac{d}{dx_b})) $

$H_{spdc} = i\hbar \chi^{(2)}(\hat{a}^{\dagger}\hat{b}\hat{b} - \hat{a}\hat{b}^{\dagger}\hat{b}^{\dagger}) =$

$-i\dfrac{\hbar \chi^{(2)}}{\sqrt{2}}(\dfrac{d}{dx_a}(\hat{x_b^2} + \dfrac{d^2}{dx_b^2}) - \hat{x_a}(\hat{1_b} + 2\hat{x}_b\dfrac{d}{dx_b}))$

#### Определение параметров

### Задаем гамильтониан SPDC

#### Первая и вторая часть MPO для когерентного состояния: $\dfrac{d}{dx_a}$ и $-\hat{x_a}$

```python
coh_operator_one = mp.der_mpo(n1, step1_v1)
coh_operator_two = mp.mpo_mul_num(mp.x_mpo(n1, x_bias - x_left, step1_v1), -1)
```

#### Первая и вторая часть MPO для вакуумного состояния: $\hat{x_b^2} + \dfrac{d^2}{dx_b^2}$ и $\hat{1_b} + 2\hat{x}_b\dfrac{d}{dx_b}$

```python
vac_operator_one = mp.mpo_sum(mp.mpo_multiplication(mp.x_mpo(n2, - x_bond2, step2), mp.x_mpo(n2, - x_bond2, step2)), mp.der2_mpo(n2, step2))
vac_operator_two = mp.mpo_sum(mp.one_mpo(n2), mp.mpo_mul_num(mp.mpo_multiplication(mp.x_mpo(n2, - x_bond2, step2), mp.der_mpo(n2, step2)), 2))
```

#### Склейка mpo для когернетного состояния и вакуумного состояния

```python
first_part = coh_operator_one + vac_operator_one
second_part = coh_operator_two + vac_operator_two
```

#### Полный гамильтониан

```python
common_operator = mp.mpo_mul_num(mp.mpo_sum(first_part, second_part), chi / (np.sqrt(2)))
```

#### Уравнение

$(1 + i\Delta t \hat{H})\psi_{t+1} = \psi_t$

#### Матрица

```python
final_mpo = mp.mpo_sum(mp.one_mpo(n1 + n2), mp.mpo_mul_num(common_operator, delta_t))
```

### Задаем гамильтониан светоделителя (для единоразового внесения потерь в моду сигнала)

$\hat{H}_{bs} = \dfrac{\theta}{2i}(\hat{1}_{pump}\hat{a}^{\dagger}_{sig}\hat{a}_{loss} - \hat{1}_{pump}\hat{a}_{sig}\hat{a}^{\dagger}_{loss}) = \dfrac{\theta}{2}\hat{1}_{pump}(\hat{q}_{sig}\hat{p}_{loss} - \hat{q}_{loss}\hat{p}_{sig}) = -i\dfrac{\theta}{2}\hat{1}_{pump}(\hat{q}_{sig}\hat{\dfrac{d}{dq}}_{loss} - \hat{q}_{loss}\hat{\dfrac{d}{dq}}_{sig})$

```python
# hamlt_1 = mp.one_mpo(n1) + mp.x_mpo(n2, -x_bond2, step2) + mp.der_mpo(n3, step3)
# hamlt_2 = mp.one_mpo(n1) + mp.der_mpo(n2, step2) + mp.x_mpo(n3, -x_bond3, step3)

hamlt_1 = mp.x_mpo(n2, -x_bond2, step2) + mp.der_mpo(n3, step3)
hamlt_2 = mp.der_mpo(n2, step2) + mp.x_mpo(n3, -x_bond3, step3)

hamiltonian = mp.one_mpo(n1) + mp.mpo_sum(hamlt_1, mp.mpo_mul_num(hamlt_2, -1))
evolution_mpo = mp.mpo_sum(mp.one_mpo(n1 + n2 + n3), mp.mpo_mul_num(hamiltonian, theta/2))
```

```python
def squeezed_vacuum(x, alpha, t):
    return np.exp(alpha * t / 2) * np.exp(- (np.exp(2 * alpha * t) * x**2) / 2) / np.pi**(0.25)
```

```python
def vacuum_state_density_matrix(x, m, omega, hbar):
    N = len(x)
    rho = np.zeros((N, N), dtype=complex)
    prefactor = (m * omega / (np.pi * hbar)) ** 0.5
    for i in range(N):
        for j in range(N):
            rho[i, j] = prefactor * np.exp(-m * omega * (x[i]**2 + x[j]**2) / (2 * hbar))
    return rho
```

```python
def find_reduced_density_matrix(mps, position, length):
    new_tensor = []
    mps_test = mps.copy()
    if position == 0:
        real_position = 1
    else:
        real_position = 0

    length2 = len(mps) - length
    frin = real_position * length
    frin2 = position * length2
    
    # Делаем свёртку тензоров MPS по физическим индексам тензоров, которые хотим исключить из рассмотрения
    for i in range(frin, frin + length2):
        new_tensor.append(np.einsum('abc, ibk -> aick', mps_test[i], np.conj(mps_test[i])))
    first = new_tensor[0]
    # Сворачиваем эти же тензоры, но теперь по бонд-индексам
    for i in range(len(new_tensor) - 1):
        first = np.einsum('abcd, cdkl -> abkl', first, new_tensor[i + 1])
    sh = first.shape
    first = first.reshape(sh[0] * sh[1], sh[2] * sh[3])
        
    start = time.time()
    # Объединяем пары тензоров нужного MPS в один тензор
    new_tensor = []
    for i in range(frin2, frin2 + length):
        t = np.einsum('inj, kml -> iknmjl', mps_test[i], np.conj(mps_test[i]))
        sh = t.shape
        t = t.reshape(sh[0] * sh[1], sh[2], sh[3], sh[4] * sh[5])
        new_tensor.append(t)
    second = new_tensor[0]
    if position == 1:
        second = np.einsum('ab, bijk -> aijk', first, second)
    #print("First part:", time.time() - start, "s")

    start = time.time()
    for i in range(length - 1):
        second = np.einsum('abcd, djkl -> abjckl', second, new_tensor[i + 1])
        sh = second.shape
        second = second.reshape(sh[0], sh[1] * sh[2], sh[3] * sh[4], sh[5])
        #print(second.shape)
        
    if position == 0:
        last = np.einsum('abcd, dk -> abck', second, first)
    if position == 1:
        last = second
    #print("Second part:", time.time() - start, "s")
    return last.reshape(last.shape[1], last.shape[2])
```

```python
def find_reduced_density_matrix2(mps, position, length):
    new_tensor = []
    mps_test = mps.copy()
    
    pos1 = position
    len1 = length
    
    pos2 = 1 - position
    len2 = len(mps) - length
    
    x1 = pos1 * len2 # Индекс начала первого mps в общем mps
    x2 = pos2 * len1 # Иднекс начала второго mps в общем mps

    # Делаем свёртку тензоров MPS по физическим индексам тензоров, которые хотим исключить из рассмотрения
    for i in range(x2, x2 + len2):
        new_tensor.append(np.einsum('abc, ibk -> aick', mps_test[i], np.conj(mps_test[i])))
    first = new_tensor[0]
    
    # Сворачиваем эти же тензоры, но теперь по бонд-индексам
    for i in range(len(new_tensor) - 1):
        first = np.einsum('abcd, cdkl -> abkl', first, new_tensor[i + 1])
    first = first.reshape(first.shape[0] * first.shape[1], first.shape[2] * first.shape[3])
   
    t1 = mps_test[x1]
    t2 = np.conj(mps_test[x1])
    for i in range(x1 + 1, x1 + len1):
        t1 = np.einsum('abc, cjk -> abjk', t1, mps_test[i])
        t1 = t1.reshape(t1.shape[0], t1.shape[1] * t1.shape[2], t1.shape[3])
        t2 = np.einsum('abc, cjk -> abjk', t2, np.conj(mps_test[i]))
        t2 = t2.reshape(t2.shape[0], t2.shape[1] * t2.shape[2], t2.shape[3])
    t = np.einsum('inj, kml -> iknmjl', t1, t2)
    t = t.reshape(t.shape[0] * t.shape[1], t.shape[2], t.shape[3], t.shape[4] * t.shape[5])
    
    if position == 0:
        last = np.einsum('abcd, dk -> abck', t, first)
    else:
        last = np.einsum('abcd, ia -> ibcd', t, first)
    
    return last.reshape(t.shape[1], t.shape[2])
```

## Моделирование спонтанного параметрического рассеяния (SPDC) с учётом потерь

### Гамильтониан SPDC с потерями:

\begin{equation*}
\begin{split}
& \hat{H} = \hat{H}_{spdc} + \hat{H}_{loss} = i\hbar \chi^{(2)}(\hat{a}_{pump}^{\dagger}\hat{a}_{sig}\hat{a}_{sig} - \hat{a}_{pump}\hat{a}_{sig}^{\dagger}\hat{a}_{sig}^{\dagger}) + \dfrac{\theta}{2i}((\hat{a}^\dagger_{pump} + \hat{a}^\dagger_{sig}) \hat{a}_{loss} - \hat{a}_{loss}^\dagger (\hat{a}_{pump} + \hat{a}_{sig})) = \\ 
& = -i\dfrac{\hbar \chi^{(2)}}{\sqrt{2}}(\dfrac{d}{dx_{pump}}(\hat{x}_{sig}^2 + \dfrac{d^2}{dx_{sig}^2}) - \hat{x}_{pump}(\hat{1}_{sig} + 2\hat{x}_{sig}\dfrac{d}{dx_{sig}})) - \dfrac{i\theta}{2}((\hat{x}_{pump} + \hat{x}_{sig})\dfrac{d}{dx_{loss}} - \hat{x}_{loss}(\dfrac{d}{dx_{pump}} + \dfrac{d}{dx_{sig}}))
\end{split}
\end{equation*}

```python
chi = 1 / 2
theta = 1
delta_t = 0.001 / 5

x_bond1 = 10
x_bond2 = 5
x_bond3 = 5
n1 = 8
n2 = 8
n3 = 8
x_bias = 20 * 5
step1 = x_bond1 / 2**(n1 - 1)
step2 = x_bond2 / 2**(n2 - 1)
step3 = x_bond3 / 2**(n3 - 1)
```

```python
coh_operartor_one = mp.der_mpo(n1, step1)
coh_operator_two = mp.mpo_mul_num(mp.x_mpo(n1, x_bias - x_bond1, step1), -1)
len(coh_operartor_one), len(coh_operator_two)
```

#### Первая и вторая часть MPO для вакуумного состояния: $\hat{x}_{sig}^2 + \dfrac{d^2}{dx_{sig}^2}$ и $\hat{1}_{sig} + 2\hat{x}_{sig}\dfrac{d}{dx_{sig}}$

```python
vac_operator_one = mp.mpo_sum(mp.mpo_multiplication(mp.x_mpo(n2, - x_bond2, step2), mp.x_mpo(n2, - x_bond2, step2)), mp.der2_mpo(n2, step2))
vac_operator_two = mp.mpo_sum(mp.one_mpo(n2), mp.mpo_mul_num(mp.mpo_multiplication(mp.x_mpo(n2, - x_bond2, step2), mp.der_mpo(n2, step2)), 2))
```

#### Склейка mpo для когернетного состояния и вакуумного состояния

```python
first_part = coh_operartor_one + vac_operator_one
second_part = coh_operator_two + vac_operator_two
```

#### Получение полного гамильтониана SPDC

```python
common_operator = mp.mpo_mul_num(mp.mpo_sum(first_part, second_part), chi / (np.sqrt(2)))
```

#### Гамильтониан потерь (модель светоделителя)

$- \dfrac{i\theta}{2}((\hat{x}_{pump} + \hat{x}_{sig})\dfrac{d}{dx_{loss}} - \hat{x}_{loss}(\dfrac{d}{dx_{pump}} + \dfrac{d}{dx_{sig}}))$

```python
p1 = mp.x_mpo(n1, x_bias - x_bond1, step1) + mp.one_mpo(n2) + mp.der_mpo(n3, step3)
p2 = mp.one_mpo(n1) + mp.x_mpo(n2, - x_bond2, step2) + mp.der_mpo(n3, step3)
p3 = mp.der_mpo(n1, step1) + mp.one_mpo(n2) + mp.x_mpo(n3, - x_bond3, step3)
p4 = mp.one_mpo(n1) + mp.der_mpo(n2, step2) + mp.x_mpo(n3, - x_bond3, step3)

p_un1 = mp.mpo_sum(p1, p2)
p_un2 = mp.mpo_mul_num(mp.mpo_sum(p3, p4), -1)
p_unsum = mp.mpo_sum(p_un1, p_un2)

p_final = mp.mpo_mul_num(p_unsum, -theta * 1j/2)
```

```python
def cut_mps(mps, n):
    n_free = n - 1
    tensor_product = mps[-1]
    first_part = mps[0][:,0,:]
    second_part = mps[0][:,1,:]
    for i in range(n_free-1):
        tensor_product = np.einsum("abc, cdf -> abdf", mps[-2-i], tensor_product)
        tensor_product = tensor_product.reshape(tensor_product.shape[0], tensor_product.shape[1]*tensor_product.shape[2], tensor_product.shape[3])
    for i in range(1, len(mps)-n+1):
        first_part = first_part @ mps[i][:,1,:]
        second_part = second_part @ mps[i][:,0,:]
    if n > 1:
        return np.concatenate((np.einsum("ab, bcd -> acd", first_part, tensor_product).reshape(-1), np.einsum("ab, bcd -> acd", second_part, tensor_product).reshape(-1)))
    else:
        return np.array((first_part, second_part)).reshape(-1)
```
