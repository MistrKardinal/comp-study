#### Функция, вычисляющая коэффициент для когерентного состояния

```python
def exponenta(x,i):
    summa = 0
    for j in range(1,i+1):
        summa += log(j)
    return exp(-(abs(x))**2 / 2 - 1 / 2 * summa + i * log(x))   
```

#### Функция, задающая коэффициенты a и b

```python
bbc, bb = sym.symbols('bbc bb')
def a_coef(n1,l1):
    nl1 = n1 - l1
    return (-1)**(nl1)*sqrt(factorial(n1)/factorial(l1))/(factorial(nl1))
def b_coef(n2,l2):
    nl2 = n2 - l2
    return sqrt(factorial(n2)/factorial(l2))/(factorial(nl2))
```

#### Интеграл от одной компоненты

```python
aac, aa = sym.symbols('aac aa')
gf = 2*pi*sym.exp(2*aa*aac)
gf = sym.sympify(gf)
def prointegr(nl1,nl2):
    s = sym.diff(gf,aa,nl1)
    return sym.diff(s,aac,nl2)
```

#### Сумма интегралов всех компонент - для конкретной пары $n$ и $n^{'}$

```python
def summa(n1,n2):
    c = 0
    for l in range(0,min(n1,n2)+1):
        c = c + a_coef(n1,l)*b_coef(n2,l)*prointegr(n1-l,n2-l)
    return c
print(summa(1,1))
```

#### Вычисление вектора когерентного состояния по заданному n

```python
import time
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(10,10))
from cmath import *
norm = 0
sr = 0
sred = 0 
ot = 0
a, b = float(input()), float(input())
x = complex(a, b)
print(x)
n = int(input())
m = []
h = []
m2 = []
numbers = []
t0 = time.perf_counter()
for i in tqdm(range(n)):
    m.append(exponenta(x,i))
#    h.append(-(abs(x))**2 / 2 - 1 / 2 * log(factorial(i)) + i * log(x))
for i in range(n):
    norm += (abs(m[i]))**2
    if (abs(m[i]))**2 >= 0.000001:    
        m2.append((abs(m[i]))**2)
        numbers.append(i)
for j in range(n):
    sr += (abs(m[j]))**2 * j
for k in range(n):
    ot += ((abs(m[k]))**2) * ((k - sr)**2)
plt.bar(numbers,m2)
print('Норма когерентного состояния при разложении по', n, 'фотонному базису', norm)
print('Среднее значение числа фотонов в когерентном состоянии', sr)
print('Среднеквадратичное стандартное отклонение в когерентном состоянии', ot)
t1 = time.perf_counter()
print(m)
print('Время выполнения кода', t1 - t0)
```

# Сжатое состояние

### Рекуррентное соотношение для полиномов Эрмита
$H_{n+1}(x) - 2xH_{n}(x) + 2nH_{n-1}(x) = 0$

```python
li = [0 for _ in range(1000)]
def hermite(x,n):
    if li[n] == 0:
        if n==0:
            li[n] = 1
        elif n==1:
            li[n] = 2*x
        else:
            li[n] = 2*x*hermite(x,n-1)-2*(n-1)*hermite(x,n-2)
    return li[n]
print(hermite(1,1))
```
