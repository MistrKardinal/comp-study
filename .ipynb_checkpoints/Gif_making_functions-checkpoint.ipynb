### Построить два изображения по двум матрицам рядом

##### Пределы осей на двойном изображении

```python
extent1 = [-(2**n1)/2 * step1 + x_bias, (2**n1)/2 * step1 + x_bias, (2**n1)/2 * step1 + x_bias, -(2**n1)/2 * step1 + x_bias]
extent2 = [-(2**n2)/2 * step2, (2**n2)/2 * step2, (2**n2)/2 * step2, -(2**n2)/2 * step2]
```

```python
def save_pair_of_matrices_as_image(matrix1, matrix2, filename, extent1, extent2, t):
    
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    
    im1 = axes[0].imshow(matrix1, cmap='viridis', extent = extent1)
    axes[0].set_title('Evolution of a coh state (pump)')
    fig.colorbar(im1, ax=axes[0])
    
    im2 = axes[1].imshow(matrix2, cmap='viridis', extent = extent2)
    axes[1].set_title('Evolution of a vac state (signal)')
    fig.colorbar(im2, ax=axes[1])
    
    n1 = int(np.log2(len(matrix1)))
    n2 = int(np.log2(len(matrix2)))
    global x_bond, x_bias, step1, step2, ph_num1, ph_num2, ph_num3
    cohn = calculate_ph_num(matrix1, n1, x_bias - x_bond, step1)
    vacn = calculate_ph_num(matrix2, n2, -x_bond, step2)
    
    plt.text(0.5, 1, f'Time {np.round(t, 4)}', transform=fig.transFigure, ha='center')
    plt.text(0.5, 1.1, f'Number of photons in coherent state {np.round(cohn, 1)}, number of photons in vacuum state {np.round(vacn, 1)}', transform=fig.transFigure, ha='center')
    plt.text(0.5, 1.05, f'sum {np.round(np.round(2 * np.round(cohn, 1), 1) + np.round(vacn, 1), 1)}', transform=fig.transFigure, ha='center')
    plt.savefig(filename, bbox_inches='tight', pad_inches=0.1)
    del im1, im2, fig, axes
    plt.close()
```
